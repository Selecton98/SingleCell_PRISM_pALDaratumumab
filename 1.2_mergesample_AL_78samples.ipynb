```r
# List of all Seurat objects
objs <- list(AL3new, AL3old, AL5new, AL5old, ALV3, ALV5, CZX3, CZX5, DGZ3, DGZ33, DGZ5, F3, F5,
             GGT3, GGT5, hg3, HG3, hg5, Hyl3, HYL3, Hyl5, LCW3, LCW5, lcwc33, lcwc35, LFM3, LFM5,
             lfy23, Lfy3, Lfy5, lxs3, LXS3, lxs5, LXS5, M3, M5, tbt3, tbt5, wdq3,
             wdq5, wjz23, wjz3, wjz5, WYQ3, WYQ5, xlg23, Xlg3, Xlg5, YCS3, YCS33, YCS35, YCS5,
             ZHZ3, ZHZ5, ZJY3, zjy5, zlj3, zlj5, zq3, zq5, zqc33, Zyg3, Zyg5, zyz3, zyz5, zyzc33, LHQ5, XFL5, ZRT5,FYD5, YYH5, D5, D3,LHQ3,XFL3,ZRT3,FYD3,YYH3)

# Extract orig.ident from meta.data
ids <- sapply(objs, function(obj) as.character(unique(obj@meta.data$orig.ident)))

# Merge all, using the extracted ids
all <- merge(objs[[1]], y = objs[-1], project = "merge", add.cell.ids = ids)
```

```r
all<- JoinLayers(all)
```

```r
options(repr.plot.height=8,repr.plot.width=25)
all[["percent.mt"]] <- PercentageFeatureSet(all, pattern = "^MT-")
all[["percent.ig"]] <- PercentageFeatureSet(all, pattern = "^IG")
all[["percent.hb"]] <- PercentageFeatureSet(all, pattern = "^HB")
VlnPlot(all, features = c("nFeature_RNA", "nCount_RNA"), ncol = 2)
VlnPlot(all, features = c("percent.mt","percent.ig","percent.hb"), ncol = 3)
```
